```python
# Dependencies
import json
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
plt.style.use('classic')
```

```python
excel_data_df = pd.read_excel('IllinoisCase%.xlsx')
ildf = pd.DataFrame(excel_data_df)
ildf.head()
```

```python
# Case by race dataframe
case_by_race_df = ildf[[ "Illinois Cases % Black",
                       'Illinois Cases % Latin']]

# Clean the data by dropping empty row
case_by_race_df = case_by_race_df.dropna(how = "any")
```

```python
crp = case_by_race_df.plot(kind='bar', color = ['darksalmon', 'midnightblue','green','magenta','goldenrod'])
plt.title('Illinois and Chicago Case Totals Compared')
plt.xlabel('Total Number of Cases')
```
