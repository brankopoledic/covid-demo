```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plot
chi_df = pd.read_csv("chicago_demo.csv",
                         keep_default_na=False, na_values=[""])
chi_df
```

```python
chicase_by_race_df = chi_df[["lab_report_date",
                      "cases_total",
                      "cases_latinx",
                      "cases_asian_non_latinx",
                      "cases_black_non_latinx",
                      "cases_white_non_latinx",
                      "cases_other_non_latinx",
                     "cases_unknown_race_eth"]]

# Clean the data by dropping empty row
chicase_by_race_df = chicase_by_race_df.dropna(how = "any")

# Change data type from str to int
cols = chicase_by_race_df.columns.drop("lab_report_date")
chicase_by_race_df[cols] = chicase_by_race_df[cols].apply(pd.to_numeric, errors='coerce')

# Remove time from date&time variable
chicase_by_race_df["lab_report_date"] = pd.to_datetime(chicase_by_race_df["lab_report_date"]).dt.date
```

```python
il_df = pd.read_csv("ILCovid.csv",
                         keep_default_na=False, na_values=[""])
il_df
```

```python
# Case by race dataframe
case_by_race_df = il_df[["Date",
                      "Cases_Total",
                      "Cases_LatinX",
                      "Cases_Asian",
                      "Cases_Black",
                      "Cases_White",
                      "Cases_AIAN",
                      "Cases_NHPI",
                       "Cases_Other"]]

# Clean the data by dropping empty row
case_by_race_df = case_by_race_df.dropna(how = "any")

# Change data type from str to int
cols = case_by_race_df.columns.drop("Date")
case_by_race_df[cols] = case_by_race_df[cols].apply(pd.to_numeric, errors='coerce')

# Remove time from date&time variable
case_by_race_df["Date"] = pd.to_datetime(case_by_race_df["Date"]).dt.date
```
