```python
import pandas as pd

covid_il = pd.read_csv("il_covid_data.csv")
covid_il.head()
```

```python
# view shape of df
covid_il.shape
```

```python
# convert date column to datetime type
covid_il["Date"] = pd.to_datetime(covid_il["Date"])
covid_il["Date"]
```

```python
# create a month column from date column
covid_il["Month"] = covid_il["Date"].dt.month_name()
covid_il.head()
```

```python
# reorder columns so month is at the front
covid_il_data = covid_il[["Month","Date", "State", "Cases_Total", "Cases_White","Cases_Black","Cases_LatinX","Cases_Asian","Cases_AIAN",
                          "Deaths_Asian","Deaths_AIAN","Deaths_NHPI","Deaths_Multiracial","Deaths_Other","Deaths_Unknown","Deaths_Ethnicity_Hispanic","Deaths_Ethnicity_NonHispanic","Deaths_Ethnicity_Unknown"]]
covid_il_data.head()
```
