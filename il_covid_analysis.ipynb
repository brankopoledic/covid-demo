```python
import pandas as pd

covid_il = pd.read_csv("csv_files/il_covid_data.csv")
covid_il.head()
```

```python
# view shape of df
covid_il.shape
```

```python
# convert date column to datetime type
covid_il["Date"] = pd.to_datetime(covid_il["Date"])
covid_il["Date"]
```

```python
# create a month column from date column
covid_il["Month"] = covid_il["Date"].dt.month_name()
covid_il.head()
```

```python
# reorder columns so month is at the front
covid_il_data = covid_il[["Month","Date", "State", "Cases_Total", "Cases_White","Cases_Black","Cases_LatinX","Cases_Asian","Cases_AIAN","Cases_NHPI","Cases_Other","Cases_Unknown",
                          "Deaths_Total",'Deaths_White',"Deaths_Black","Deaths_LatinX","Deaths_Asian","Deaths_AIAN","Deaths_NHPI","Deaths_Other","Deaths_Unknown"]]
covid_il_data.head()
```

```python
# get count of null values and location
covid_il_data.isnull().sum()
```

```python
# normalize data 
state_population = 12600
white_cases_per_100 = (covid_il_data["Cases_White"]/covid_il_data["Cases_Total"]) * 100
white_cases_per_100 
```

```python
# group data by month
march_data = covid_il_data.loc[covid_il_data["Month"] == "March", :]
april_data = covid_il_data.loc[covid_il_data["Month"] == "April", :]
may_data = covid_il_data.loc[covid_il_data["Month"] == "May", :]
june_data = covid_il_data.loc[covid_il_data["Month"] == "June", :]
july_data = covid_il_data.loc[covid_il_data["Month"] == "July", :]
```

```python
import matplotlib.pyplot as plt
```

```python
# total cases bar graph

y_value = "Total Cases"

month_and_total_cases = covid_il_data[["Month","Cases_Total"]]
month_and_total_cases = month_and_total_cases.set_index("Month")

month_and_total_cases.plot(kind="bar", figsize=(20,3))
plt.title("Total Covid Cases in IL by Month")
plt.ylabel(f"{y_value}")
plt.show();
```

```python
# total Deaths bar graph

y_value = "Total Deaths"

month_and_total_cases = covid_il_data[["Month","Deaths_Total"]]
month_and_total_cases = month_and_total_cases.set_index("Month")

month_and_total_cases.plot(kind="bar", figsize=(20,3))
plt.title("Total Covid Deaths in IL by Month")
plt.ylabel(f"{y_value}")
plt.show();
```
