```python
import pandas as pd

il_pop = pd.read_csv('csv_files/Illinois_demographics.csv')

il_pop.head(50)
```

```python
# look at shape
il_pop.shape
```

```python
il_pop.columns = map(str.lower, il_pop.columns)
il_pop.columns = ['id_race', 'race', 'id_eth', 'ethnicity', 'id_year', 'year',
       'hispanic_pop', 'geography', 'id_geography',
       'slug_geography', 'il_population_2018', 'share']
il_pop
```

```python
# create new df for just 2018 data
il_pop_2018 = il_pop.loc[il_pop["id_year"] == 2018, :]
il_pop_2018
```

```python

il_pop_2018 = il_pop.loc[il_pop['id_year'] == 2018,["race","ethnicity","il_population_2018","share"]]
il_pop_2018 ["race/ethnicity"] = ["black","latinx","white","latinx","latinx","other","latinx","other","other","latinx","latinx","asian","other","latinx"]
il_pop_2018
```

```python
il_race = il_pop_2018.groupby('race/ethnicity')
race_count_il = pd.DataFrame(il_race['il_population_2018'].sum())
race_count_il
```

```python
il_ethnicity = il_pop_2018.groupby('ethnicity')
ethnicity_count_il = pd.DataFrame(il_ethnicity['il_population_2018'].sum())
ethnicity_count_il
```

```python
race_count_il.to_csv("csv_files/il_race_count.csv")
```
